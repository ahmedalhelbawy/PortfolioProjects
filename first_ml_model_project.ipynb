## R**andom Forest**

###**Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

###**Applying the model to make prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

###**Ecaluate model perfomence**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame([['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]])
rf_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
rf_results
```

# Data Visualization of prediction results

```python
#Normal Visualization

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5, 5))
plt.scatter(x=y_train, y=y_train_pred, c="#7CAE00", alpha=0.3)

z = np.polyfit(y_train, y_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
```

```python
#Professional Visualization

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.scatter(y_train, y_train_pred, c="#7CAE00", alpha=0.6, edgecolors='k', s=100)

z = np.polyfit(y_train, y_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), color='#F8766D', linewidth=2)

plt.title('Predicted vs. Experimental LogS', fontsize=16)
plt.xlabel('Experimental LogS', fontsize=14)
plt.ylabel('Predicted LogS', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```
